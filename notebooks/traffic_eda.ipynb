```python
import matplotlib.pyplot as plt
```

```python
df["Traffic Situation"].value_counts().plot(
    kind="bar",
    color="skyblue"
)

plt.title("Traffic Situation Distribution")
plt.show()
```

```python
vehicle_cols = [
    "CarCount",
    "BikeCount",
    "BusCount",
    "TruckCount"
]

df[vehicle_cols].mean().plot(
    kind="bar",
    color="orange"
)

plt.title("Average Vehicle Counts")
plt.show()
```

```python
df["Time"] = pd.to_datetime(df["Time"])

df["Hour"] = df["Time"].dt.hour

hourly = df.groupby("Hour")["Total"].mean()

hourly.plot(figsize=(10,5))
plt.title("Average Traffic by Hour")
plt.show()
```
